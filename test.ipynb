```python
import pickle as pkl
```

```python
path = "data/ADNI/beta/ad/002_S_1018_I60267_MP-RAGE_REPEAT_beta.pkl"
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
img = pkl.load(open(path, "rb"))
```

```python
# show 100 slices of the image in the middle
fig = plt.figure(figsize=(10, 10))
for i in range(30, 130):
    ax = fig.add_subplot(10, 10, i - 29)
    ax.imshow(img[i, :, :], cmap="gray")
    ax.axis("off")
plt.tight_layout()
plt.show()
```
