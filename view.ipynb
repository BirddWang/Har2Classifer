```python
import nibabel as nib
import numpy as np
import matplotlib.pyplot as plt
import pickle as pkl
```

```python
dir = "data/ADNI"
name = "002_S_0619_I16390_MP-RAGE_REPEAT"
```

```python
pre_img = nib.load(f"{dir}/preprocessed/ad/{name}_prep.nii.gz")
beta_img = pkl.load(open(f"{dir}/beta/ad/{name}_beta.pkl", "rb"))
harmonized_img = nib.load(f"{dir}/harmonized/ad/{name}_prep_harmonized_fusion.nii.gz")
```

```python
print(pre_img.get_fdata().shape)
print(beta_img.shape)
print(harmonized_img.get_fdata().shape)
```

```python
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(pre_img.get_fdata()[100, :, :], cmap="gray")
plt.axis("off")
plt.subplot(1, 3, 2)
slice = beta_img[100, :, :]
slice = np.rot90(slice, 1)  # Rotate the slice for better visualization
# 
slice = np.flipud(slice)
plt.imshow(slice, cmap="gray")
plt.axis("off")
plt.subplot(1, 3, 3)
plt.imshow(harmonized_img.get_fdata()[100, :, :], cmap="gray")
plt.axis("off")
plt.tight_layout()
plt.show()
```
